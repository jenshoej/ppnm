```python
import numpy as np
import matplotlib.pyplot as plt


ts = np.loadtxt("second-order-test.txt", usecols=0)
y0 = np.loadtxt("second-order-test.txt", usecols=1)
y1 = np.loadtxt("second-order-test.txt", usecols=2)

plt.figure(dpi = 150)
plt.plot(ts, y0, label="y0")
plt.plot(ts, y1, label="y1")
plt.legend()
plt.show()
```

```python
t = np.loadtxt("SIR.txt", usecols=0)
S = np.loadtxt("SIR.txt", usecols=1)
I = np.loadtxt("SIR.txt", usecols=2)
R = np.loadtxt("SIR.txt", usecols=3)

plt.figure(dpi = 150)
plt.plot(t, S, label="S", color="blue")
plt.plot(t, I, label="I", color="red")
plt.plot(t, R, label="R", color="green")
plt.legend()
plt.show()
```
